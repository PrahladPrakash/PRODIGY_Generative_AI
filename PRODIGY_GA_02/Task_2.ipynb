```python
# Import necessary libraries
from diffusers import StableDiffusionPipeline
import torch
```

```python
# Load the Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")

# Check for CUDA availability and move model if available
if torch.cuda.is_available():
    pipe.to("cuda")  # Ensure you're using the GPU for faster inference if available
    print("Model moved to CUDA")
else:
    print("CUDA not available, using CPU") # Inform the user if CUDA is not available and the model will run on CPU
```

```python
# Define your text prompt
prompt = "A serene mountain landscape with a crystal-clear lake and a vibrant sunset"
```

```python
# Generate the image
image = pipe(prompt).images[0]

# Display the generated image
image.show()

# Save the generated image
image.save("generated_image.png")
```
